```python
import pandas as pd
#1: Loading the data
df = pd.read_csv("/content/Sample raw dataset.csv")
print("Original dataset\n",df)
```

```python
#2: Check missing values
print("Missing values", df.isnull().sum())
```

```python
#3: Drop missing value
df_drop_rows = df.dropna()
print("Drop rows:\n",df_drop_rows)

df_drop_columns = df.dropna(axis=1)
print("Drop columns:\n",df_drop_columns)
```

```python
# Fill with constants
df_fill_constant= df.fillna({'Age':0,'Gender':'Unknown','Purchase_Amount':0})
print("After filling with constants\n",df_fill_constant)
```

```python
#Aggregation
df_mean = df.copy()
df_mean['Age'] = df_mean['Age'].fillna(df_mean['Age'].mean())
print("After filling with mean\n",df_mean)
df_mean['Purchase_Amount'] = df_mean['Purchase_Amount'].fillna(df_mean['Purchase_Amount'].mean())
print("After filling with mean\n",df_mean)

df_mean['Gender'] = df_mean['Gender'].fillna(df_mean['Gender'].mode()[0])
print("After filling with mean\n",df_mean)

df_median = df.copy()
df_median['Age'] = df_median['Age'].fillna(df_median['Age'].median())
print("After filling with median\n\n",df_median)
df_median['Purchase_Amount'] = df_median['Purchase_Amount'].fillna(df_median['Purchase_Amount'].median())
print("After filling median for purchase amount\n\n",df_median)


```

```python
# Forward Fill and Backward fill
df_ffill = df.fillna(method = 'ffill')
print("After forward fill\n",df_ffill)

df_bfill = df.fillna(method = 'bfill')
print("After backward fill\n",df_bfill)
```

```python
#Interpolation
df_interpolated = df.interpolate(method = 'linear')
#print("After interpolation\n",df_interpolated)

#combined Vertically
df_combined = pd.concat([df_mean,df_interpolated],axis = 0)
df_combined.to_csv('combined.csv',index = False)
print("Combined dataset\n",df_combined)

#combine horizontally
df_combined = pd.concat([df_mean,df_interpolated],axis = 1)
df_combined.to_csv('combined_horizontally.csv',index = False)
print("Combined Horizontally\n",df_combined)

#combine with tagging
df_interpolated_copy = df_interpolated.copy()
df_mean_copy = df_mean.copy()

df_interpolated_copy['Method'] = 'Y'
df_mean_copy['Method'] = 'X'

df_combined_tag = pd.concat([df_mean_copy,df_interpolated_copy])
df_combined_tag.to_csv('combined_tag.csv',index = False)
```
