#Labels and title

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x, y)
plt.title("Sports Watch Data",loc = 'right',fontdict = font1)
plt.xlabel("Average Pulse",fontdict = font2)
plt.ylabel("Calorie Burnage",fontdict = font2)
plt.grid()

plt.show()
```

#Histogram
A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(80, 5, 40)    #values concentrate around 80(marks), std deviation 5(group) and total value is 40(no of students)
print(x)
plt.hist(x)
plt.show()
```

#Bins


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (random normal distribution)
data = np.random.normal(loc=50, scale=15, size=1000)

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Histogram with bins=5
axs[0].hist(data, bins=5, color='skyblue', edgecolor='black')
axs[0].set_title("Histogram with 5 Bins")

# Histogram with bins=10
axs[1].hist(data, bins=10, color='mediumseagreen', edgecolor='black')
axs[1].set_title("Histogram with 10 Bins")

# Histogram with bins=20
axs[2].hist(data, bins=20, color='salmon', edgecolor='black')
axs[2].set_title("Histogram with 20 Bins")

# Shared styling
for ax in axs:
    ax.set_xlabel("Value")
    ax.set_ylabel("Frequency")
    ax.grid(True)

plt.suptitle("Effect of Changing Bins in Histogram", fontsize=16)
plt.tight_layout()
plt.show()
```

#Pie chart
