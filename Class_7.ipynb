```python
import pandas as pd
import matplotlib.pyplot as plt


#1 Load data
df = pd.read_csv("/content/netflix.csv")
#print(df.head())

#2 Remove duplicates
df = df.drop_duplicates()
#print(df.head())

####

#3 Handle missing values
#replace director with "Not available"
df['director'] = df['director'].fillna("Not available")
#replace country with "Not Specified"
df['country'] = df['country'].fillna("Not Specified")
#fill missing date_added with "Unknown"
df['date_added'] = df['date_added'].fillna("Unknown")
#Missing rating with "Unrated"
df['rating'] = df['rating'].fillna("Unrated")

#4 handling inconsistencies
df['date_added']=pd.to_datetime(df['date_added'],errors = 'coerce')
print(df.head())

#Transformation of Data
#1 extract year and the month
df['added_year'] = df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.month
print(df.head())

#2 extracting duration number and duration type
df['duration_num'] = df['duration'].str.extract('(\d+)').astype(int)
df['duration_type'] = df['duration'].str.extract('([A-Za-z]+)').astype(str)
print(df.head())

# Clean white spaces
text_cols = ['title','director','country','listed_in']
for col in text_cols:
  df[col] = df[col].str.strip()

print(df.head())

#Split multivalue column
df['listed_in_list'] = df['listed_in'].str.split(',')
print(df.head())
#iloc[]

df.to_csv('netflix_cleaned.csv',index=False)
print("Clean file Created")

#number of titles added per year

#split as per genre

#study content based trend

#study country trend
country_trend = df.groupby(['added_year','type']).size().unstack(fill_value = 0)
country_trend.plot(figsize = (12,6))
plt.title("Country wise content trend")
plt.show()

```
