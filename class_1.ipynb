```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/fitness_data_150_entries.csv")
df.head(10)

print("\n The shape of the dataset(row & Column):",df.shape)
print("\n The summary of the dataset:\n",df.describe())
```

```python
#Access a particular column
print(df['Calories'])
```

```python
#Access multiple columns
print(df[['Duration','Pulse']].head())
```

```python
#Accesing rows by index
print(df.iloc[10:15])
```

```python
#Filter data
#conditional stmts
Long_workout = df[df['Duration']>45]
print(Long_workout)
```

```python
high_calories = df[df["Calories"]>340]
print(high_calories)
```

```python
#Aggregation
print(df['Calories'].mean())
```

```python
#Group by
grouped = df.groupby('Duration')['Pulse'].mean()
print(grouped)
```

```python
#Sort the data
print(df.sort_values(by='Calories',ascending=False))
```

```python
#Add new feature  -  Feature engineering
df['Calories_per_min'] = df['Calories']/df['Duration']
print(df.head())
```

```python
#Save csv
df.to_csv("fitness_data_updated.csv")
print(df.head())
```
