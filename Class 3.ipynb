#Tuples


```python
x =("Pen",1,"Pencil")
print(x)
x = ("Pen", 1, 24.6, "Female")
print(x)
print(x[1])
print(x[-1])
print(x[0:3])
print(x[-2:])
```

#Packing elements

```python
x = ("Paris","Rome","Italy")
(a,b,c) = x
print(a,b,c)
print(x)
print(b)

x = ("Paris","Rome","Italy","Florence","Venice")
(a,*b,c,d) = x
print(a,b,c)
print(b)
```

#unpack

```python
x = ("Paris","Rome",("Italy","Florence","Venice"))
(a,b,(c,d,e)) = x
print(a,b,c,d,e)
print(a,b,c)
print(d)
```
